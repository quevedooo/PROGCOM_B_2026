###Estructura de repetición
1. for= Cuando se sabe cuantas veces se debe repetir una acción
2. while. No se sabe cuantas veces se repite la acción




##1. FOR
Sintaxis general en python

```
for (variable) in (algo):
 se repite una acción
```
algo
* in range(n): quiero repetir n veces
* in range (CI,CF,p): Quiero repetir desde CI, hasta CF-1, con paso p
* in (lista, diccionario): Cuando quiero iterar cada elemento de la lista/diccionario
* each (lista, diccionario): Cuando quiero iterar cada elemento dentro d ela lista/diccionario


````python
#Quiero repetir n veces
for i in range(9):
  print(f"hay {i+1} personas en el salón")
````

````python
for i in range(1,10):
  print(f"hay {i} personas en el salón")
````

````python
dias=int(input("Ey!! Cuantos días falta para tu cumpleaños!?"))
for cumple in range(dias,0,-1):
  print(f"Faltan{cumple} días para tu cumpleaños!!")
print("FELIZ CUMPLEAÑOS!😘😎😜")
````

#WHILE

Sintaxis

```
while (CI):
 lo que se va a repetir
 modificador de CI
```
**OJO!!**
- while arranca si CI es VERDADERA
- modificadores de CI: pregunta,contador,acumulador,cambio de variable




````python
c1=0
c2=0
c3=0
c4=0
nulo=0
voto=int(input("Ingrese el número de su candidato (1-4)"))
if voto==1: c1=c1+1
elif voto==2: c2=c2+1
elif voto==3: c3=c3+1
elif voto==4: c4=c4+1
else: nulo=nulo+1
print("Votos para el primer canditato:",c1)
print("Votos para el segundo canditato:",c2)
print("Votos para el tercer canditato:",c3)
print("Votos para el cuarto canditato:",c4)
print("Votos nulos:",nulo)
````
