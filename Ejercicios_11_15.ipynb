##CALCULADORA SIMPLE

```python
num1 = float(input("Ingrese el primer número: "))
num2 = float(input("Ingrese el segundo número: "))
operación = input("Ingrese la operación (+, -, *, /): ")
if operación == "+":
    resultado = num1 + num2
elif operación == "-":
    resultado = num1 - num2
elif operación == "*":
    resultado = num1 * num2
elif operación == "/":
    if num2 != 0:
        resultado = num1 / num2
    else:
        resultado = "Error: no se puede dividir entre cero."
else:
    resultado = "Operación no válida."

print("Resultado:", resultado)
```

##CALIFICACION EN LETRAS

```python
nota = int(input("Ingrese la nota (0 - 100): "))
if nota >= 90 and nota <= 100:
    letra = "A"
elif nota >= 80:
    letra = "B"
elif nota >= 70:
    letra = "C"
elif nota >= 60:
    letra = "D"
elif nota >= 0:
    letra = "F"
else:
    letra = "Nota inválida."
print("Su calificación es:", letra)
```

##CALIFICACIÓN EN LETRAS IF TERNARIO

```python
nota = int(input("Ingrese la nota (0 - 100): "))
letra = (
    "A" if 90 <= nota <= 100 else
    "B" if 80 <= nota < 90 else
    "C" if 70 <= nota < 80 else
    "D" if 60 <= nota < 70 else
    "F" if 0 <= nota < 60 else
    "Nota inválida."
)
print("Calificación:", letra)
```

##VALIDACION DE TRIANGULO

```python
lado1 = float(input("Ingrese el primer lado: "))
lado2 = float(input("Ingrese el segundo lado: "))
lado3 = float(input("Ingrese el tercer lado: "))
if lado1 + lado2 > lado3 and lado1 + lado3 > lado2 and lado2 + lado3 > lado1:
    if lado1 == lado2 and lado2 == lado3:
        tipo = "Triángulo equilátero."
    elif lado1 == lado2 or lado1 == lado3 or lado2 == lado3:
        tipo = "Triángulo isósceles."
    else:
        tipo = "Triángulo escaleno."
else:
    tipo = "No forman un triángulo."
print("Resultado:", tipo)
```
