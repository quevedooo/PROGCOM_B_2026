##Suma de números naturales

```python
##FOR
suma=0
n=int(input("Ingrese el valor de n:"))
for i in range(1,n+1):
  suma+=i
print("La suma de los números naturales es:",suma)
```

```python
##WHILE
suma=0
n=int(input("Ingrese el valor de n:"))
i=1
while i<=n:
  suma+=i
  i+=1
print("La suma de los números es:",suma)
```

##Factorial de un número

```python
##FOR
numero=int(input("Ingrese su número:"))
if numero<0:
  print("Los números negativos no tienen factorial")
else:
  factorial=1
  for i in range(1,numero+1):
    factorial *=i
  print("El factorial de su número es:",factorial)
```

```python
##WHILE
numero=int(input("Ingrese su número:"))
if numero<0:
  print("Los números negativos no tienen factorial")
else:
  factorial=1
  i=1
  while i <=numero:
    factorial*=i
    i+=1
  print("El factorial de su número es:",factorial)
```

##Números de Fibonacci

```python
##FOR
n = int(input("Ingrese la cantidad de números de Fibonacci: "))
a = 0
b = 1
for i in range(n):
    print(a, end=" ")
    siguiente = a + b
    a = b
    b = siguiente
```

```python
##WHILE
n = int(input("Ingrese la cantidad de números de Fibonacci: "))
a = 0
b = 1
contador = 0
while contador < n:
    print(a, end=" ")
    siguiente = a + b
    a = b
    b = siguiente
    contador += 1
```

##Números pares del 1 al 100

```python
##FOR
for numero in range(1, 101):
    if numero % 2 == 0:
        print(numero)
```

```python
##WHILE
numero = 1
while numero <= 100:
    if numero % 2 == 0:
        print(numero)
    numero += 1
```

##Suma de los dígitos de un número

```python
##FOR
numero = input("Ingrese un número: ")
suma = 0
for digito in numero:
    suma += int(digito)
print("La suma de los dígitos es:", suma)
```

```python
##WHILE
numero = int(input("Ingrese un número: "))
suma = 0
while numero > 0:
    digito = numero % 10
    suma += digito
    numero //= 10
print("La suma de los dígitos es:", suma)
```

##Determinar si un número es primo

```python
##FOR
numero = int(input("Ingrese un número: "))
contador = 0
for i in range(1, numero + 1):
    if numero % i == 0:
        contador += 1
if contador == 2:
    print("El número es primo")
else:
    print("El número no es primo")
```

```python
##WHILE
numero = int(input("Ingrese un número: "))
i = 1
contador = 0
while i <= numero:
    if numero % i == 0:
        contador += 1
    i += 1
if contador == 2:
    print("El número es primo")
else:
    print("El número no es primo")
```

##Patrón de triángulo con asteriscos

```python
##FOR
n = int(input("Ingrese la altura del triángulo: "))
for i in range(1, n + 1):
    print("*" * i)
```

```python
##WHILE
n = int(input("Ingrese la altura del triángulo: "))
i = 1
while i <= n:
    print("*" * i)
    i += 1
```

##Calcular potencia sin usar **

```python
##FOR
base = int(input("Ingrese la base: "))
exponente = int(input("Ingrese el exponente: "))
resultado = 1
for i in range(exponente):
    resultado *= base
print("El resultado de la potencia es:", resultado)
```

```python
##WHILE
base = int(input("Ingrese la base: "))
exponente = int(input("Ingrese el exponente: "))
resultado = 1
i = 0
while i < exponente:
    resultado *= base
    i += 1
print("El resultado de la potencia es:", resultado)
```
