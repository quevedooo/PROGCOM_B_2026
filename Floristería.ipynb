```python
color_techo="blanco"
color_pared="blanco"
cant_ventanas=3
cant_puertas=1
cant_pisos=1
cant_baños=1
color_marco_ventanas="negro"
decolarión_flores=True
flores_decoración="Hojas y flores"
letrero=True
nombre_letrero=""
grafiti=True
dibujo_grafiti="Diablo y homero"
```

```python
def decorando_floristería(a,b,c,d,e,f,g,h,letrero,grafiti):
 color_techo=a
 color_pared=b
 cant_ventanas=c
 cant_puertas=d
 cant_pisos=e
 cant_baños=f
 color_marco_ventanas=g
 decolarión_flores=h
 if h:flores_decoración=input("Ingrese las flores con las que decoró el lugar")
 if letrero:nombre_letrero=input("Ingrese el nombre de su letrero")
 if grafiti:dibujo_grafiti=input("Ingrese el dibujo de su grafiti")
 print(f"Descripción de mi floreistería:\nEl color del techo es{a}, \nEl color de la pared es{b}, \nLa cantidad de ventanas es{c}, \nLa cantidad de puertas es{d}, \nLa cantidad de pisos es{e}, \nLa cantidad de baños es{f}, \n El color del marco de las ventanas es{h} ")
```
