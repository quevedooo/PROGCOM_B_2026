````python
figura=["angelitos","estrella fugaz", "dinosaurios","diablitos"]
cantidad=[4,2,5,3]
````

##Operaciones con listas
1. Recorrer listas
2. Verificar elementos dentro de la lista
3. Borrar elementos en la lista
4. Agregar elementos a la lista
5. Mutabilidad de listas


###1. Recorrer listas


````python
#len:Calcula el tamaño de un dato
for elemento in range(len(figura)):
  print(figura[elemento])
````

````python
for elemento in range(len(figura)):
  print(figura[elemento],":",cantidad[elemento])
````

````python
for elemento in figura:
  print(elemento)
````

###2. Verificar elementos dentro de la lista


````python
user=input("Qué figura quiere comprar?")
for elemento in range(len(figura)):
  if user==figura[elemento]: print("Si hay🍪")
  else: print("No hay 😢")
````

````python
if user in figura: print("Si hay🍪")
else: print("No hay 😢")
````
