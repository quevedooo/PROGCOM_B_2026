Programa tabla de multiplicar

```python
##for
print("La tabla del 4 es:")
for i in range(1,11):
 print(f"4*{i}: {i*4}")
```

```python
##While
print("La tabla del 4 es:")
i=1
while i<=10:
  print(f"4*{i}= {i*4}")
  i+=1
```
